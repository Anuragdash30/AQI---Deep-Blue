```python
wcheddha=pd.read_csv('CheddaNagar.csv')
wcheddha.head()
```

```python
wtilak=pd.read_csv('TilakNagar.csv')
wsindhi=pd.read_csv('SIndhiSociety.csv')
wmahul=pd.read_csv('MahulApi.csv')
wchembur=pd.read_csv('Chembur_west.csv')

aqw= pd.concat(
    [wcheddha, wtilak, wsindhi, wmahul, wchembur],
    axis=1,
    join="inner",
    ignore_index=False,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
   
)
aqw.head()
```

```python
dfavg=aqw['AQI'].mean(axis=1, skipna=True)
```

```python
dfavg.head()
```

```python
newaqv= aqw[['Date', 'AQI']]
newaqv.columns=['date', 'date1', 'date2','date3', 'date4', 'aqicheddha','aqitilak','aqisindhi', 'aqimahul', 'aqichembur']
newaqv.head()
```

```python
aqve=newaqv.drop(['date1', 'date2', 'date3', 'date4'], axis=1)
aqve.head()
```

```python
aqvf=aqve.dropna(axis=0, subset=['date'])
```

```python
west = aqvf.assign(avg=aqvf.iloc[:,1:7].mean(axis=1))
west.head()
```

```python
west.to_csv('westaqi.csv')
```
