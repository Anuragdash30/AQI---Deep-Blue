```python
wmankhurd=pd.read_csv('MankhurdWest.csv')
wmankhurd.head()
```

```python
wdeonar=pd.read_csv('Deonar.csv')
wcheeta=pd.read_csv('CheetaCampApi.csv')
wanus=pd.read_csv('AnushaktiApi.csv')
wshiv=pd.read_csv('ShivajiNagar.csv')
wgovandi=pd.read_csv('GovandiE.csv')
wcheme=pd.read_csv('ChemburE.csv')
wtrom=pd.read_csv('TrombayApi.csv')


aqw= pd.concat(
    [wmankhurd, wdeonar, wcheeta, wanus, wshiv, wgovandi, wcheme, wtrom],
    axis=1,
    join="inner",
    ignore_index=False,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
   
)
aqw.head()
```

```python
dfavg=aqw['AQI'].mean(axis=1, skipna=True)
```

```python
dfavg.head()
```

```python
newaqv= aqw[['Date', 'AQI']]
newaqv.columns=['date', 'date1', 'date2','date3', 'date4','date4','date4','date4', 'aqiman', 'aqideonar', 'aqicheeta', 'aqianush', 'aqishiv', 'aqigov', 'aqicheme', 'aqitrom']
newaqv.head()
```

```python
aqve=newaqv.drop(['date1', 'date2', 'date3', 'date4'], axis=1)
aqve.head()
```

```python
aqvf=aqve.dropna(axis=0, subset=['date'])
```

```python
east = aqvf.assign(avg=aqvf.iloc[:,1:7].mean(axis=1))
east.head()
```
